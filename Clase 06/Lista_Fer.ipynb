CLASE NODO

```python
class Node:
    def __init__(self,data):
        self.__data = data
        self.__next = None

    def getData(self):
        return self.__data
        
    def getNext(self):
        return self.__next

    def setData(self, data):
        self.__data = data

    def setNext(self, next):
        self.__next = next
```

CLASE LISTA

```python
class LinkedList:

    def __init__(self):
        self.__head = None

    def setHead(self, item):
        self.__head = item
    
    def getHead(self):
        return self.__head
    
    def isEmpty(self):
        return (self.__head == None)
    
    def addItem(self, item):
        """Add item to the end of the list"""
        new_node = Node(item)
        new_node.setNext(self.getHead)
        self.setHead = new_node
        return None
    
    def size(self):
        count = 0
        current = self.getHead()
        while(current != None):
            count  += 1
            current = current.getNext()
        return count
    
    def search(self, item):
        """Search for item in list. If found, return True. If not found, return False"""
        current = self.getHead()
        while(current != None):
            if(current.getData() == item):
                return True
            else:
                current = current.getNext()
        return False
    
    def remove(self, item):
        previous = None
        current = self.getHead()

        while(current != None):
            if(current.getData() == item):
                if(previous == None):
                    self.setHead(None)
                else:
                    previous.setNext(current.getNext())
                break
            else:
                previous = current
                current = current.getNext()

        return None
    
    def insert(self,position,item):
        
        if (position > (self.size() - 1)):
            print('Index out of bounds')
            raise IndexError
        if (position == 0):
            self.addItem(item)
        else:
            previous = None
            current = self.getHead()
            current_position = 0
            while(current_position < position):
                previous = current
                current = current.getNext()
                current_position += 1
            new_node = Node(item)
            previous.setNext(new_node)
            new_node.setNext(current)
        return None

    def indexOf(self,item):
        current_position = 0
        current = self.getHead()

        while(current != None):
            if(current.getData() == item):
                return current_position
            else:
                current_position += 1
                current = current.getNext()

        return None

    def pop(self, position=None):
        """
        If no argument is provided, return and remove the item at the head. 
        If position is provided, return and remove the item at that position.
        If index is out of bounds, raise IndexError
        """
        if (self.isEmpty()):
            return None
        
        current = self.getHead()
        
        if(position == None):
            self.setHead(current.getNext())
            return current.getData()
        else:
            if(position > (self.size() -1)):
                print('Index out of bounds')
                raise IndexError
            current_position = 0
            previous = None

            while(current_position < position):
                previous = current
                current = current.getNext()
                current_position += 1

            previous.setNext(current.getNext())
            return current.getData()
        
    def append(self, item):
        size = self.size()
        current_position = 0
        current = self.getHead()
        
        new_node = Node(item)

        if(current == None):
            self.setHead(new_node)
        else:
            while(current_position < size):
                current = current.getNext()
                current_position += 1
            current.setNext(new_node)

        return None

    def printList(self):
        current = self.getHead()
        while(current != None):
            print(current.getData())
            current = current.getNext()
```
