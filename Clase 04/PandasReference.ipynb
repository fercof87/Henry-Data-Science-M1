AGRUPACIONES

```python
import pandas as pd

# Creamos un diccionario con mucha información
ventas = {
    'Comercial': ['Juan', 'María', 'Manuel', 'Vanesa', 'Ana', 'Marcos'],
    'Empresa': ['Movistar', 'Jazztel', 'Movistar', 'Jazztel', 'Vodafone', 'Vodafone'],
    'Primas': [300, 220, 140, 70, 400, 175]
} 
```

```python
df = pd.DataFrame(ventas)
df
```

```python
# Utilizando `groupby` podemos agrupar las filas en función del nombre de la columna, 
# por ejemplo el nombre de la `Empresa`, al hacerlo se generará un nuevo objeto de tipo 
# `DataFrameGroupBy`:
por_empresa = df.groupby("Empresa")
```

```python
# Prima media por empresa
por_empresa.mean("prima")
```

```python
df.groupby('Empresa').mean("prima")
```

```python
# Primas mínimas (error)
por_empresa.min()
```

```python
# ID de las primas mínimas
por_empresa['Primas'].idxmin()
```

```python
# Usamos las ID de las primas máximas como fuente del df
df.loc[por_empresa['Primas'].idxmin()]
```

```python
# Reporte de analíticas descriptivas por empresa
por_empresa.describe()
```

COMBINACION DE DATAFRAMES

CONCATENACION

```python
#Consiste en juntar filas de varios DF, deben coincidir las dimensiones de los DF
df1 = pd.DataFrame({'A':['1A','2A'], 'B':['1B','2B'], 'C':['1C','2C']}, index=[1, 2]) 
df2 = pd.DataFrame({'A':['3A','4A','5A'], 'B':['3B','4B','5B'], 'C':['3C','4C','5C']}, index=[3, 4, 5]) 
df3 = pd.DataFrame({'A':['6A','7A'], 'B':['6B','7B'], 'C':['6C','7C']}, index=[6, 7]) 
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
# Concatenar filas
pd.concat([df1, df2, df3])
```

FUSION ==> MERGE

```python
# Esta técnica permite fusionar los `DataFrames` utilizando la misma lógica que se utiliza en `SQL` 
# para unir tablas, por ejemplo una combinación de dos dataframes a partir 
# de una columna común:
df1 = pd.DataFrame({'clave': ['K1','K2','K3'], 'A': ['1A','2A','3A'], 'B': ['1B','2B','3B']})
df2 = pd.DataFrame({'clave': ['K1','K2','K3'], 'C': ['1C','2C','3C'], 'D': ['1D','2D','3D']})
```

```python
df1
```

```python
df2
```

```python
# Fusionar columnas a partir de una columna común
pd.merge(df1, df2, on='clave')
```

UNION ==> JOIN

```python
# Esta técnica es muy conveniente para combinar las columnas de dos `dataframes` indexados:
# En vez de buscar una columna comun, usamos el INDICE COMUN
df1 = pd.DataFrame({'A': ['1A','2A','3A'], 'B': ['1B','2B','3B']}, index=['K1','K2','K3'])
df2 = pd.DataFrame({'C': ['1C','2C','3C'], 'D': ['1D','2D','3D']}, index=['K1','K2','K3'])
```

```python
df1
```

```python
df2
```

```python
# Unir columnas mediante los índices
df1.join(df2)
```

WEB SCRAPING

```python
# Realizamos un scrapping de una tabla de la wikipedia
df = pd.read_html('https://web.archive.org/web/20220717170349/https://en.wikipedia.org/wiki/List_of_countries_by_past_fertility_rate')
```

```python
# selecciono la tabla que quiero, porque en este caso guarde mas de una
df[2]
```

```python
# Guardamos el dataframe
fertility_rate = df[2]

fertility_rate.head()
```

```python
# Renombramos la primera columna para que sea más fácil consultarla
fertility_rate.rename(columns = {'Country/dependent territory':'Country'}, inplace=True)

fertility_rate
```

```python
# Índice de natalidad por país entre los años 2010-2015
fertility_rate[["Country", "2010–2015"]]
```

```python
# Misma consulta aplicando el styler para esconder la primera columna
fertility_rate[["Country", "2010–2015"]].head().style.hide(axis=0)
```

```python
# Índice de natalidad por país entre los años 1985–1990 ordenado de más a menos (primeros resultados)
fertility_rate[["Country", "1985–1990"]].sort_values(by="1985–1990", ascending=False).head().style.hide(axis=0)
```

```python
# Índice de natalidad por país entre los años 1985–1990 ordenado de más o menos (últimos resultados)
fertility_rate[["Country", "1985–1990"]].sort_values(by="1985–1990", ascending=False).tail().style.hide(axis=0)
```

```python
# Vamos a transformar todas las columnas desde la segunda hasta la última a valores númericos
fertility_rate = fertility_rate[1:][:].apply(pd.to_numeric, errors='coerce')
```

```python
fertility_rate
```

```python
# Ahora podemos consultar la media del índice de natalidad para cada año
fertility_rate.mean()[1:]
```

```python
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = 10,5

fertility_rate.mean()[1:].plot(kind='line', xlabel="Períodos", ylabel="Media de natalidad mundial")
```

```python
import pandas as pd

# Iterar utilizando iterrows()
data = {'Nombre': ['Juan', 'María', 'Pedro'],
        'Edad': [25, 30, 35]}
df = pd.DataFrame(data)

for indice, fila in df.iterrows():
    print("Índice:", indice)
    print("Fila:")
    print(fila)
```

```python
# Iterar utilizando itertuples() ==> MAS OPTIMO!!!!!
for fila in df.itertuples():
    print("Índice:", fila.Index)
    print("Nombre:", fila.Nombre)
    print("Edad:", fila.Edad)
```
